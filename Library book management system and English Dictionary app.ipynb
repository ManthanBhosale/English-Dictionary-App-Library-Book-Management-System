# English Dictionary app

This project uses the skills learnt from week 1 to week 6. The aim of this project is to create an English Dictionary app that returns definitions of English words. When the application is started, it displays a menu as follows:

Main Menu
Add a new word

Find the meaning

Update a word

Exit

If item 1 is selected, it should prompt for a word and then accept its meaning. This pair of word and its meaning should be stored in a file called ‘words.txt’. When item 2 is selected, it should prompt for a word, search that word in the file ‘words.txt’ and return the meaning if it is found. If the meaning is not found, it should display an appropriate message. Similarly, when choice 3 is entered, it should accept a word and update its meaning. In all the above three cases, the prompt should return to the main menu. The words and their meanings should be stored as a dictionary in words.txt, using the concept of serialization. If choice 4 is selected, a graceful exit should be performed.

```python
#~~~~~~~~~~~~~import libraries~~~~~~~~~~
import json
from IPython.display import clear_output
from functools import reduce
import os

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
def main():
    """Main method to call function based on choice"""
    while True:
        clear_output(wait=True)
        n = int(input('\nEnter choice in number.\n1.ADD:\n2.Find:\n3.Update:\n0.Exit\n'))
        clear_output(wait=True)
        if n == 1:
            add()
        elif n == 2:
            find()
        elif n == 3:
            update()
        elif n == 0:
            Exit()
            break
        else:
            print('Please enter correct input:')

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~            
def insert(line,word,meaning):
    with open('words.txt','w') as f:
        if line.startswith('{') == True and line.endswith('}') == True:
            f.write(line.replace('}',','))
            f.write('"'+word+'":"'+meaning+'"}')
        else:
            f.write('{"'+word+'":"'+meaning+'"}') 
        print('Added Successfully!')

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
def add():
    #~~~~~~~~~~~~~~~~~~~~~Clearing Scree
    clear_output(wait=True)
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~READING words.txt FILE
    if 'words.txt' in os.listdir():
        with open('words.txt','r') as f:
            line = f.readline()
    else:
        with open('words.txt','w+') as f:
            line = f.readline()
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~INPUT WORD AND MEANING
    word = input(('Enter word to Add in the Dictionary: '))
    
    if len(line) != 0: 
        
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Serialization    
        liner = line.replace("'",'"')
        fword = json.loads(liner)
        if word not in fword:
            meaning = input(('Enter Meaning of the word: '))
            word = word.replace(' ','')
        
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~~WRITING words.txt FILE with WORD AND MEANING
            insert(line,word,meaning)
        else:
            print('Word already exist in the Dictionary!')
    else:
        meaning = input(('Enter Meaning of the word: '))
        insert(line,word,meaning)
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~READING words.txt FILE After Writing with new Word and Meaning
    with open('words.txt','r') as f:
        line = f.readline()
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~replace " with ' for the last element
    line = line.replace('"',"'")
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Converting DATA from words.txt to dict() format
    line_r = line.replace("'",'"')
    line1 = json.loads(line_r)
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Converting the dict() type data into str() format
    line2 = line1.copy()
    line2 = str(line2)
    with open('words.txt','w') as f:
        f.write(line2)
    
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
def find():
    clear_output(wait=True)
    
    word = input('Enter the word you want to search : ')
    
    with open('words.txt','r') as f:
        line = f.readline()
    line_r = line.replace("'",'"')
    line1 = json.loads(line_r)
    
    for key,value in line1.items():
        if key == word:
            print(key)
            print(value+'.')
            return
        else:
            continue
    print('Word not found!')
            
def update():
    
    clear_output(wait=True)
    with open('words.txt','r') as f:
        line = f.readline()
   
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~Serialization    
    liner = line.replace("'",'"')
    fword = json.loads(liner)    
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~INPUT WORD AND MEANING 
    word = input(('Enter the word to Add in the Dictionary: '))   
    word = word.replace(' ','')
    if word in fword.keys():
        meaning = input(('Enter the Meaning of the word: '))
        fword[word] = meaning
    else:
        print('Word not found !. Try Adding the word.')
    
    final_txt = fword.copy()
    final_txt = str(final_txt)
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~write final updated word and meaning in words.txt file
    final_txt = final_txt.replace('"',"'")
    with open('words.txt','w') as f:
        f.write(final_txt)

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
def Exit():
    print('Thank for using English Dictionary App, see you soon!')
    return
main()
```

# Library Book Management System

1B: Library Book Management System

In this project, a Library Book Management System will be implemented with the front end GUI developed using Tkinter and the backend database implemented in SQLite. It is a single window tkinter GUI application which will implement CRUD( create, read, update, delete) operations on a ‘books’ database, in SQLite. The window will contain text widgets to accept / display book title, author, year of publication, ISBN (optional) and a large text area/ list box to display the records which match the search criteria.

```python
#========================================================LIBRARIES================================================================
from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox

root = Tk()
root.title('LIBRARY BOOK MANAGEMENT APPLICATION') #title for the application

#===========================================================================================================================
def Database():
    """Database() creates a connection for database file Books.db, creates if doesnt exist, and creates a table booksinfo if not exist"""
    global conn, cursor 
    conn  = sqlite3.connect('Books.db')
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS booksinfo(booktitle TEXT,author TEXT,year TEXT,isbn TEXT)") #creating table booksinfo

#===========================================================================================================================

def Add():
    """Add() method inserts books informaion into the database"""
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Retrieving text from Textbox
    TITLE = str(booktitle.get("0.0",END+"-1c"))
    AUTHOR = str(author.get("0.0",END+"-1c"))
    YEAR = str(year.get("0.0",END+"-1c"))
    ISBN = str(isbn.get("0.0",END+"-1c"))
    
    ##~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Comparing if the any text box is empty
    if TITLE == "" or AUTHOR == "" or YEAR == "" or ISBN == "":
        txt_result.config(text = "PLEASE ENTER ALL FIELDS!",fg='red')
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Inserting retrieved data in the table
    else:
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Checking if data already exist
        Database()
        cursor.execute("SELECT * FROM booksinfo")
        conn.commit()
        fetch = cursor.fetchall()
        element = (TITLE,AUTHOR,YEAR,ISBN)
        for data in fetch:
            if data == element:
                txt_result.config(text="Data already exist!",fg="red")
                booktitle.delete("1.0",END)
                author.delete("1.0",END)
                year.delete("1.0",END)
                isbn.delete("1.0",END)
                return
            
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Insert data in database
        cursor.execute("INSERT INTO booksinfo values(?,?,?,?)",(TITLE,AUTHOR,YEAR,ISBN))
        conn.commit()
        booktitle.delete("1.0",END)
        author.delete("1.0",END)
        year.delete("1.0",END)
        isbn.delete("1.0",END)
        cursor.close()
        conn.close()
        txt_result.config(text="SUCCESSFULLY ADDED DATA",fg="green")

#===========================================================================================================================

def Search():
    """Search() method searches the book information specified"""
    
    #~~~~~~~~~~~~~~~~~~~~Clearing the existing data displayed~~~~~~~~~~~~~~~~~~~~
    
    for item in tree.get_children():
        tree.delete(item)
        
    #~~~~~~~~~~~~~~~~~~~~~~~~~~Getting data from textbox~~~~~~~~~~~~~~~~~~~~~~~~~
    
    TITLE = str(booktitle.get("0.0",END+"-1c"))
    AUTHOR = str(author.get("0.0",END+"-1c"))
    YEAR = str(year.get("0.0",END+"-1c"))
    ISBN = str(isbn.get("0.0",END+"-1c"))
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Checking if any textbox empty~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
    
    if TITLE == "" or AUTHOR == "" or YEAR == "" or ISBN == "":
        txt_result.config(text = "Please enter all the fields",fg='red')
    else:
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~retrieving entered data from database~~~~~~~~~~~~~~~~~~~~~~~~~
    
        Database()
        cursor.execute("SELECT * FROM booksinfo where booktitle = ? and author = ? and year = ? and isbn = ?",(TITLE,AUTHOR,YEAR,ISBN)) #ask
        conn.commit()
        fetch = cursor.fetchall() 
        
    #~~~~~~~~~~~~~~~~~~~~~~~~~if data doesnt exist displaying message~~~~~~~~~~~~~~~~~~~~~~~~~~~
    
        if len(fetch) == 0:
            txt_result.config(text = "Data not found!",fg = "black")
            
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~displaying the exist data~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
        
        else:
            for data in fetch:
                tree.insert('','end',values=(data[0],data[1],data[2],data[3])) 
            cursor.close()
            conn.close()
            booktitle.delete("1.0",END)
            author.delete("1.0",END)
            year.delete("1.0",END)
            isbn.delete("1.0",END)    
            txt_result.config(text = "Successfully fetched data !",fg = "black")
            
#===========================================================================================================================

def EXIT():
    """EXIT() method exits the window"""
    #~~~~~~~~~~~~~~~~~~~~~~~~~~messagebox, asking for confirmation to exit~~~~~~~~~~~~~~~~~~~~~~~~~`
    result = tkMessageBox.askquestion('Do you want to exit ? (y/n)',icon = 'warning') 
    if result == 'yes':
        root.destroy()
    
#=========================================================FRAME================================================================
Top = Frame(root, width=900,height=50, bd=8, relief ='raise')
Top.pack(side=TOP)

Left = Frame(root,width=300,height=500,bd=8,relief ='raise')
Left.pack(side=LEFT)

Right = Frame(root,width=600,height=500,bd=8,relief ='raise')
Right.pack(side=RIGHT)

Forms = Frame(Left,width=300,height=450)
Forms.pack(side=TOP)

Buttons = Frame(Left,width=300,height=100,bd=8,relief='raise')
Buttons.pack(side=BOTTOM)

#===============================================LABEL WIDGET===================================================================
txt_title=Label(Top,width=900,text='A Simple CRUD Application')
txt_title.pack(side=LEFT)
txt_booktitle=Label(Forms, text="BookTitle:",font=('arial',16),bd=15)
txt_booktitle.grid(row=0,stick="w")
txt_author=Label(Forms, text="Author:",font=('arial',16),bd=15)
txt_author.grid(row=1,stick="w")
txt_year=Label(Forms, text="Year:",font=('arial',16),bd=15)
txt_year.grid(row=3,stick="w")
txt_isbn=Label(Forms, text="ISBN:",font=('arial',16),bd=15)
txt_isbn.grid(row=4,stick="w")
txt_result=Label(Buttons)
txt_result.pack(side=TOP)
#=================================================TEXT WIDGET===================================================================
booktitle = Text(Forms, height = 1,width=30)
booktitle.grid(row=0,column=1)
author = Text(Forms, height = 1,width=30)
author.grid(row=1,column=1)
year = Text(Forms, height = 1,width=30)
year.grid(row=3,column=1)
isbn = Text(Forms, height = 1,width=30)
isbn.grid(row=4,column=1)
#==================================================BUTTONS WIDGET===================================================================
btn_add = Button(Buttons,width=10,text="Add",command=Add)
btn_add.pack(side = LEFT)

btn_search = Button(Buttons,width=10,text='Search',command=Search)
btn_search.pack(side = LEFT)

#btn_update = Button(Buttons,width=10,text='Update',command=Update)
#btn_update.pack(side = LEFT)

#btn_delete = Button(Buttons,width=10,text='Delete',command=Delete)
#btn_delete.pack(side = LEFT)

btn_exit = Button(Buttons,width=10,text='Exit',command=EXIT)
btn_exit.pack(side = LEFT)

#=================================================LIST WIDGET======================================================================

scrollbary = Scrollbar(Right, orient = VERTICAL)
scrollbarx = Scrollbar(Right, orient =  HORIZONTAL)

tree = ttk.Treeview(Right, columns=("BookTitle","Author","Year","ISBN"))
scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT,fill=Y)
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM,fill=X)

tree.heading('BookTitle',text="Book Title",anchor=W)
tree.heading('Author',text="Author",anchor=W)
tree.heading('Year',text="Year",anchor=W)
tree.heading('ISBN',text="ISBN",anchor=W)
tree.column('#0',stretch=NO,minwidth=80,width=0)
tree.column('#1',stretch=NO,minwidth=0,width=200)
tree.column('#2',stretch=NO,minwidth=0,width=180)
tree.column('#3',stretch=NO,minwidth=0,width=80)
tree.column('#4',stretch=NO,minwidth=0,width=100)
tree.pack()
#=================================================INITIALIZATION======================================================================

if __name__=='__main__':
    root.mainloop()
```
